```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import plotly.graph_objects as go
from sklearn.preprocessing import LabelEncoder
import io

# Outputs
output_file = widgets.Output()
output_head = widgets.Output()
output_exploratory = widgets.Output()
output_plot1 = widgets.Output()

#Segundo gráfico
output_plot2 = widgets.Output()

# Widgets
file_selector = widgets.FileUpload(accept='.csv,', multiple=False) 
#file_selector = widgets.FileUpload(accept='.csv', multiple=False)
button_load = widgets.Button(description="Cargar Dataset")

# Variables globales
df = pd.DataFrame()
df2 = pd.DataFrame()
Dx = widgets.Dropdown(description='Variable X')
Dy = widgets.Dropdown(description='Variable Y')

#Crea los widgets x categoricas y y numericas
Dx2 = widgets.Dropdown(description='Categorica')
Dy2 = widgets.Dropdown(description='Numerica')

Op = widgets.FloatSlider(
    value=0.5,
    min=0,
    max=1,
    step=0.01,
    description='Opacidad:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
)

# Función para crear gráficos
def create_plots(x, y):
    with output_plot1:
        output_plot1.clear_output()
        fig = go.FigureWidget(data=[go.Scatter(x=df[x], y=df[y], mode='markers')])
        fig.update_layout(title=f'Dispersión: {x} vs {y}')
        display(fig)

def create_plots2(X, Y, z):
    with output_plot2:
        output_plot2.clear_output()
        # Crear el gráfico de violín con Plotly Express
        fig2 = go.Figure()
        
        fig2.add_trace(go.Violin(
            x=df2[X],
            y=df2[Y],
            box_visible=True,
            points='all',
            opacity=z
        ))

        fig2.update_layout(title=f'Gráfico de Violín: {X} y {Y}')
        display(fig2)

# Ejemplo de uso:
# create_violin_plot("ocean_proximity", "households", 0.5, df2)



# Función para análisis exploratorio
def exploratory_analysis():
    with output_exploratory:
        output_exploratory.clear_output()
        print("Descripción del dataset:")
        display(df.describe())
        print("\nDistribución de valores nulos:")
        display(df.isnull().sum())

# Función para cargar el dataset
def load_data(b):
    global df, df2
    #output_file.clear_output()
    # Espera a que el usuario suba un archivo
    #Cambiar upload por fileselector
    while not file_selector.value:
        pass
    # Obtiene el contenido del archivo
    uploaded_file = file_selector.value[0]
    file_name = uploaded_file['name']
    file_data = uploaded_file['content']
    df = pd.read_csv(pd.io.common.BytesIO(file_data), sep= ";")
    df2 = df.copy()
    # Codificar variables categóricas
    for column in df.select_dtypes(include=['object']).columns:
        df[column] = LabelEncoder().fit_transform(df[column])
    
    if file_selector.value:
        try:
            with output_file:
                print("✅ Datos cargados exitosamente")

            # Mostrar primeras filas
            with output_head:
                output_head.clear_output()
                display(df.head())

            # Análisis exploratorio
            exploratory_analysis()

            # Configurar dropdowns
            Dx.options = df.columns.tolist()
            Dx.value = df.columns[0]
            Dy.options = df.columns.tolist()
            Dy.value = df.columns[1]

            # Seleccionar variables numéricas
            Num = df2.select_dtypes(include=['number']).columns.tolist()

            # Seleccionar variables categóricas
            NoNum = df2.select_dtypes(include=['object', 'category']).columns.tolist()
            
            Dx2.options = NoNum
            Dx2.value = NoNum[0]
            Dy2.options = Num
            Dy2.value = Num[0]


            # Activar interacción
            widgets.interact(create_plots, x=Dx, y=Dy)

            widgets.interact(create_plots2, X=Dx2, Y=Dy2, z = Op)

            

        except Exception as e:
            with output_file:
                print(f"Error al cargar el archivo: {e}")
    else:
        with output_file:
            print("Por favor selecciona un archivo .csv antes de cargar.")

# Asociar botón
button_load.on_click(load_data)

# Crear pestañas
tab = widgets.Tab()
tab.children = [
    widgets.VBox([file_selector, button_load, output_file]),
    output_head,
    output_exploratory,
    widgets.VBox([Dx, Dy, output_plot1]),
    widgets.VBox([Dx2, Dy2, Op, output_plot2])
]
tab.set_title(0, 'Seleccionar y Cargar Datos')
tab.set_title(1, 'Primeras Filas')
tab.set_title(2, 'Análisis Exploratorio')
tab.set_title(3, 'Scatter Plot')
tab.set_title(4, 'Violin')

# Mostrar interfaz
display(tab)
```
